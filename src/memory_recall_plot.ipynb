```python
batches = ['batch01', 'batch02', 'batch03', 'batch05', 'batch06', 'batch07', 'batch08', 'batch09', 'batch10', 'batch11']
Rlist_names = ['R_f01_L10', 'R_f01_L20', 'R_f01_L50', 'R_f01_L80', 'R_f01_L130', 
               'R_f01_L280', 'R_f01_L500', 'R_f005_L10', 'R_f005_L20', 'R_f005_L50', 
               'R_f005_L80', 'R_f005_L130', 'R_f005_L280', 'R_f005_L500','R_f001_L10',
               'R_f001_L20', 'R_f001_L50', 'R_f001_L80', 'R_f001_L130', 'R_f001_L280', 
               'R_f001_L500']

R_f01_L10 =  []
R_f01_L20 =  []
R_f01_L50 =  [] 
R_f01_L80 =  []
R_f01_L130 = []
R_f01_L280 = []
R_f01_L500 = [] 

R_f005_L10 =  []
R_f005_L20 =  []
R_f005_L50 =  []
R_f005_L80 =  []
R_f005_L130 = []
R_f005_L280 = []
R_f005_L500 = [] 

R_f001_L10 =  []
R_f001_L20 =  []
R_f001_L50 =  []
R_f001_L80 =  []
R_f001_L130 = []
R_f001_L280 = []
R_f001_L500 = []

Rlist = [R_f01_L10, R_f01_L20, R_f01_L50, R_f01_L80, R_f01_L130, R_f01_L280, 
         R_f01_L500, R_f005_L10, R_f005_L20, R_f005_L50, R_f005_L80, R_f005_L130, 
         R_f005_L280, R_f005_L500, R_f001_L10, R_f001_L20, R_f001_L50, R_f001_L80, 
         R_f001_L130, R_f001_L280, R_f001_L500]

for i in range(len(batches)):
    for j in range(len(Rlist_names)):
        with open('/content/'+batches[i]+'/'+Rlist_names[j]+'.pkl', 'rb') as f:
            loaded_list = pickle.load(f)
            Rlist[j] += loaded_list
            f.close()
```

```python
R_f01_mean  = [np.mean(R_f01_L10), 
               np.mean(R_f01_L20), 
               np.mean(R_f01_L50),
               np.mean(R_f01_L80),
               np.mean(R_f01_L130), 
               np.mean(R_f01_L280), 
               np.mean(R_f01_L500)]

R_f005_mean  = [np.mean(R_f005_L10), 
                np.mean(R_f005_L20), 
                np.mean(R_f005_L50),
                np.mean(R_f005_L80),
                np.mean(R_f005_L130), 
                np.mean(R_f005_L280), 
                np.mean(R_f005_L500)]

R_f001_mean  = [np.mean(R_f001_L10), 
                np.mean(R_f001_L20), 
                np.mean(R_f001_L50),
                np.mean(R_f001_L80),
                np.mean(R_f001_L130), 
                np.mean(R_f001_L280), 
                np.mean(R_f001_L500)]
```

```python
f_val = [0.1, 0.05, 0.01]
L_val = [10, 20, 50, 80, 130, 280, 500]

def TheoreticalScalingZeroSparseness(L):
    return np.sqrt(1.5*np.pi*float(L))

def PowerLaw(x, alpha, pre):
    return pre * x**alpha
```

```python
par_R_f01, cov_R_f01   = curve_fit(f=PowerLaw, xdata=L_val, ydata=R_f01_mean, p0=[0.41, 2.5])
par_R_f005, cov_R_f005 = curve_fit(f=PowerLaw, xdata=L_val, ydata=R_f005_mean, p0=[0.41, 2.5])
par_R_f001, cov_R_f001 = curve_fit(f=PowerLaw, xdata=L_val, ydata=R_f001_mean, p0=[0.41, 2.5])

print('Fit f=0.1')
print('- alpha =', par_R_f01[0])
print('- prefactor =', par_R_f01[1])
print('=========================================')
print('Fit f=0.05')
print('- alpha =', par_R_f005[0])
print('- prefactor =', par_R_f005[1])
print('=========================================')
print('Fit f=0.01')
print('- alpha =', par_R_f001[0])
print('- prefactor =', par_R_f001[1])

fit_alpha = [par_R_f01[0], par_R_f005[0], par_R_f001[0]]
fit_pre = [par_R_f01[1], par_R_f005[1], par_R_f001[1]]
```

```python
colors = [u'#1f77b4', u'#ff7f0e', u'#2ca02c', u'#d62728', u'#9467bd', u'#8c564b', u'#e377c2', u'#7f7f7f', u'#bcbd22', u'#17becf']

if latex == True:
    plt.figure(figsize=(10,7))
    plt.xlabel("List length", fontsize=22)
    plt.ylabel("Retrieved items (mean)", fontsize=22)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.grid()
    rlist = [R_f01_mean, R_f005_mean, R_f001_mean]
    plt.plot(L_val, [TheoreticalScalingZeroSparseness(val) for val in L_val], '--', c=colors[4], label='$\sqrt{3\pi L/2}$')
    for r in range(3):
        plt.plot(L_val, rlist[r], 'o', c=colors[r], label='$f = '+str(f_val[r])+'$') 
        plt.plot(L_val, PowerLaw(L_val, fit_alpha[r], fit_pre[r]), '-', c=colors[r])
    plt.legend(fontsize=18)
    plt.savefig('r_vs_l.pdf', bbox_inches='tight')
    plt.show()
elif latex == False:
    plt.figure(figsize=(10,7))
    plt.xlabel("List length", fontsize=22)
    plt.ylabel("Retrieved items (mean)", fontsize=22)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.grid()
    rlist = [R_f01_mean, R_f005_mean, R_f001_mean]
    plt.plot(L_val, [TheoreticalScalingZeroSparseness(val) for val in L_val], '--', c=colors[4], label='Theory for f~0')
    for r in range(3):
        plt.plot(L_val, rlist[r], 'o', c=colors[r], label='f = '+str(f_val[r])) 
        plt.plot(L_val, PowerLaw(L_val, fit_alpha[r], fit_pre[r]), '-', c=colors[r])
    plt.legend(fontsize=18)
    plt.savefig('r_vs_l.pdf', bbox_inches='tight')
    plt.show()
```
